```python
import seaborn as sns 
from matplotlib import pyplot as plt
import pandas as pd
sns.set_context("poster")
sns.axes_style()
sns.despine()
from scipy import stats
```

```python
df = pd.read_csv('data.csv')
print(df.head(5))
```

```python
df
```

```python
sns.lmplot(x='frame', y='x_position', data=df,
           fit_reg=True)
 
```

```python
# get coeffs of linear fit
slope, intercept, r_value, p_value, std_err = stats.linregress(df[:20]['frame'],df[:20]['position'])

# use line_kws to set line label for legend
ax = sns.regplot(x="frame", y="x_position", data=df[:20],
 line_kws={'label':"y={0:.1f}x+{1:.1f}".format(slope,intercept)})

# plot legend
ax.legend()

sns.despine()
plt.show()
```

```python
ax.get_figure().savefig("output.png")
```
